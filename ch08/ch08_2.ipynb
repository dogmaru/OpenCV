```python
from google.colab import drive
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt

drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/scanImg.png')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gImg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg.shape)
#plt.imshow(gImg, cmap='gray')
cv2_imshow(gImg)
```

```python
kernel = np.ones((4,1), np.uint8)  #구조요소의 모양 적절하게 변경
dilateImg = cv2.dilate(gImg, kernel, iterations = 1)
cv2_imshow(dilateImg)  #실제로 프로그램 상에선 흰색이 오브젝트이므로 침식연산을 하면 검은색 부분이 확장됨. 따라서 검은색 부분 없애기 위해 팽창연산 적용
```

```python
kernel = np.ones((4,1), np.uint8)
erodeImg = cv2.erode(dilateImg, kernel, iterations = 1)
cv2_imshow(erodeImg)
```
