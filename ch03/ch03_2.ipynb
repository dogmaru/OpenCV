```python
from google.colab import drive
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt

drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/girl.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gImg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg.shape)
cv2_imshow(gImg)
```

```python
gPlus = np.zeros((gImg.shape[0], gImg.shape[1]), dtype = np.ubyte) #ubyte: unsigned byte(0~255)
```

```python
val = 20
```

```python
for h in range(gImg.shape[0]):
  for w in range(gImg.shape[1]):
    plusVal = gImg[h,w] + val
    if (plusVal == 255):   #클램핑
      plusVal = 255
    gPlus[h, w] = plusVal

plt.imshow(gPlus, cmap = 'gray')
```

```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/aero2.bmp')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gImg1 = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg1.shape)
cv2_imshow(gImg1)
```

```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/lena256.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gImg2 = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg2.shape)
cv2_imshow(gImg2)
```

```python
#평균 연산

for h in range(gImg.shape[0]):
  for w in range(gImg.shape[1]):
    #plusVal = np.int32(gImg1[h,w]) + np.int32(gImg2[h,w])
    plusVal = (np.int32(gImg1[h,w]) + np.int32(gImg2[h,w])) / 2   #평균내기
    if (plusVal == 255):   #클램핑
      plusVal = 255
    gPlus[h, w] = plusVal

cv2_imshow(gPlus)
```

```python
#페이드인/페이드아웃 구현

weight = 0.1

for h in range(gImg.shape[0]):
  for w in range(gImg.shape[1]):
    plusVal = np.int32(gImg1[h,w]) * weight + np.int32(gImg2[h,w]) * (1 - weight)  #각 이미지에 가중치 두기
    if (plusVal == 255):   #클램핑
      plusVal = 255
    gPlus[h, w] = plusVal

cv2_imshow(gPlus)
```
